# Data Cleaning

```python
# Getting information on what are the data types of each column in the dataframe

airbnb_Boston.dtypes
```

#### Now here we convert the last_review column has float64 dtype let's change that to datetime format 

```python
# Using the to_datetime(yourdf['column']) method
airbnb_Boston['last_review'] = pd.to_datetime(airbnb_Boston['last_review'])
```

```python
#Display the dataframe
airbnb_Boston.head(30)
```

```python
#From the datetime series we wish to extract the year from it and make a new column year

airbnb_Boston['year'] = airbnb_Boston['last_review'].dt.year
airbnb_Boston['year']
```

##### Using the strip method we can remove the unwanted whitespaces from the string data

```python
#airbnb_Boston['name'] = airbnb_Boston['name'].str.strip()
airbnb_Boston['name'] = airbnb_Boston['name'].str.strip()
```

```python
airbnb_Boston
```

```python
airbnb_Boston['name']
```

#### Since, the name does not have a standardized capitalization in the name column using .str.lower() method to lowercase every letter

```python
airbnb_Boston['name-lower'] = airbnb_Boston['name'].str.lower()
```

```python
# Just to see the that the name data has been turned to lowercase letters
airbnb_Boston[['name-lower', 'name']]
```

```python
# Now we will calculate the minimum revenue generated by calculating the minimum number of nights along with the price generated

airbnb_Boston['min_revenue'] = airbnb_Boston['minimum_nights']*airbnb_Boston['price']

airbnb_Boston[['minimum_nights', 'price', 'min_revenue']].head()

#airbnb_Boston.plot.scatter( x= 'minimum_nights' , y = 'price', )
```

#### Once the data cleaning is done we move to descriptive analysis where some questions like what is the median pricing for renting homes at airbnb in boston. 

```python
#To do so we calculate mean of the price data using the df.mean() method

print("The average pricing in Boston is :" + str(airbnb_Boston['price'].mean()))

```

```python
#similarly we calculate median using df.median() method

airbnb_Boston['price'].median()
```

#### Here Mean is greater than Median hence the data is right skewed and hence we will rely on median for effective distribution

```python
# Now, if we want to analyse typical data or prices on the bases of type of house we can use df.groupby() method 

airbnb_Boston[['room_type','price']].groupby('room_type',as_index = False).mean()
```

#### We can see here that what a typical price in airbnb in boston usually is respective to the type of house. 

```python
#Similarly we calculate median 

median_price = airbnb_Boston[['room_type','price']].groupby('room_type',as_index = False).median()
median_price
```

##### Now I decided to do some fun plotting the graphs, so skip the next cell if you wish to

#### Plotting a scatter plot

```python
median_price.plot.scatter(x='room_type' , y= 'price', title = 'Scatter plot of Median of Price vs Room_Type')
```

```python
median_price.plot.box(x='room_type' , y= 'price', title = 'Boxplot of Median of Price vs Room_Type')
```

### Obervation : 
From the above graph we can see that the Entire home/apt have highest median prices followed by Hotel room and the lowest being the shared apartments

## To get a better output I decided to groupby the median of prices on the basis of years and the type of apartments

```python
yearly_med_price = airbnb_Boston[['room_type', 'year', 'price']].groupby(['room_type', 'year'], as_index = False).median()
yearly_med_price
```

### Obersvation:  
From the above table it can be observed that over the years the median prices in Entire Home/Apt have been fluctuating, whereas in hotel rooms and private rooms have decreased over the years, however for shared rooms they have increased 

### Filtering Data

Now since, I am looking for apartments in boston below 100$ amount since that is the most common price among the listings hence I will use a boolean expression to filter out such data

```python
# Total areas in Boston using df.unique() function I will get a array in which every unique area in Boston which has airbnb will be printed out 

airbnb_Boston['neighbourhood'].unique()
```

```python
airbnb_in_Bos_under100 = airbnb_Boston[airbnb_Boston['price'] < 100]
airbnb_in_Bos_under100.head()
```

### Applying more filters!

### Now, I want to see the apartments in Back Bay area in the year 2021 with price below 100$

```python
# To do so we apply using expressions in paranthesis
#get all rows in back bay area with price below 100$ in 2021

airbnb_Boston_under100_BB_2021  = airbnb_Boston[(airbnb_Boston['price'] < 100) & ((airbnb_Boston['year'] == 2021)) & (airbnb_Boston['neighbourhood'] == 'Back Bay')]

airbnb_Boston_under100_BB_2021.head()
```

```python
airbnb_Boston_BB_2021  = airbnb_Boston[(airbnb_Boston['price'] < 100)& (airbnb_Boston['year'] == 2021) & (airbnb_Boston['neighbourhood'] == 'Back Bay')]
airbnb_Boston_BB_2021.head()
```
